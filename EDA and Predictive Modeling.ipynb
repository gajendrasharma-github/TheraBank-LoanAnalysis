```python
from sklearn.model_selection import train_test_split
x=data.drop(columns=['Personal Loan'])
y=data['Personal Loan']
xtrain,xtest, ytrain, ytest=train_test_split(x,y, train_size=0.7, random_state=1)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

#### To check the distribution in y_train and y_test is same as y we will use chi-square goodness of fit

```python
expected= y.value_counts(normalize=True).values
expected
```

```python
obs_train= ytrain.value_counts(normalize=True)
```

```python
obs_test=ytest.value_counts(normalize=True)
```

```python
import scipy.stats as stats
```

```python
#H0: The distribution is same
#H1: The distribution is not same

stats.chisquare(obs_train, expected)
```

```python
#since pvalue>0.05 , we fail to reject the null hypothesis, distribution is same
```

```python
stats.chisquare(obs_test, expected)
```

```python
#since pvalue>0.05 , we fail to reject the null hypothesis, distribution is same
```

### Model significance
- LLRT (Log Likelihood Ratio Test to check the significance of the model)
- H0: The model is not significant
- HA: The model is significant

```python
llr=-2*(base_model.llnull-base_model.llf)
```

```python
llr
```

```python
base_model.llr
```

```python
# We see this statistic in Chi Square Distribution with Degree of freedom=df of model
```

```python
stats.chi2.sf(llr, 10)
```

```python
base_model.llr_pvalue
```

```python
# Since pvalue is less than 0.05, hence model is significant
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import roc_auc_score, roc_curve, recall_score, precision_score, f1_score, accuracy_score
```

```python
models=[]
accuracy=[]
precison=[]
recall=[]
f1=[]
```

```python
def model_validation(model, xtrain, ytrain, xtest, ytest):
    m=model
    m.fit(xtrain, ytrain)
    pred=m.predict(xtest)
    prob=m.predict_proba(xtest)[:,1]
    
    print('confusion_matrix: \n', confusion_matrix(ytest, pred))
    print('Classification Report :\n', classification_report(ytest, pred))
    print('ROC AUC: ', roc_auc_score(ytest,prob))
    
    fpr, tpr,_=roc_curve(ytest, prob)
    plt.plot(fpr, tpr, color='green')
    plt.plot([0,1],[0,1], color='red')
    plt.show()
    
    resp=input('Do you want to Save this model in summary? Y/N')
    if resp.lower()=='y':
        models.append(str(m))
        accuracy.append(accuracy_score(ytest, pred))
        recall.append(recall_score(ytest, pred))
        precison.append(precision_score(ytest, pred))
        f1.append(f1_score(ytest,pred))
        global model_df
        model_df= pd.DataFrame({'Model':models, 'Accuracy': accuracy, 'Recall': recall, 'Precison':precison, 'F1':f1})
    else:
        return
    
    base_model=sma.Logit(ytrain, xtrain).fit()
base_model.summary()
```

```python
def model_validation(model, xtrain, ytrain, xtest, ytest):
    m=model
    m.fit(xtrain, ytrain)
    pred=m.predict(xtest)
    prob=m.predict_proba(xtest)[:,1]
    
    print('confusion_matrix: \n', confusion_matrix(ytest, pred))
    print('Classification Report :\n', classification_report(ytest, pred))
    print('ROC AUC: ', roc_auc_score(ytest,prob))
    
    fpr, tpr,_=roc_curve(ytest, prob)
    plt.plot(fpr, tpr, color='green')
    plt.plot([0,1],[0,1], color='red')
    plt.show()
    
    resp=input('Do you want to Save this model in summary? Y/N')
    if resp.lower()=='y':
        models.append(str(m))
        accuracy.append(accuracy_score(ytest, pred))
        recall.append(recall_score(ytest, pred))
        precison.append(precision_score(ytest, pred))
        f1.append(f1_score(ytest,pred))
        global model_df
        model_df= pd.DataFrame({'Model':models, 'Accuracy': accuracy, 'Recall': recall, 'Precison':precison, 'F1':f1})
    else:
        return
    
    base_model=sma.Logit(ytrain, xtrain).fit()
base_model.summary()
gscv=GridSearchCV(estimator=GradientBoostingClassifier(), param_grid=params, cv=5, verbose=1, scoring='f1')
final_model=GradientBoostingClassifier(**gscv.best_params_)
```
